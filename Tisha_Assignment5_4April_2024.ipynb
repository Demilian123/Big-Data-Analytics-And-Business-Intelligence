```python
# Q.1 Apply seaborn library to do the following

# 1. Read Total profit of all months and show it using a line plot.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv')

# Plotting data
plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=data, marker='o')
plt.title('Total Profit Over Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
# 2. Read all product sales data and show it using a multiline plot.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv')

# Plotting data
plt.figure(figsize=(10, 6))
for product in data.columns[1:7]:
    sns.lineplot(x='month_number', y=product, data=data, label=product, marker='o')
plt.title('Product Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Plot bar chart for all the features/attributes.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv')

# Plotting data
plt.figure(figsize=(12, 6))
sns.barplot(data=data.drop('month_number', axis=1), errorbar=None)
plt.title('Bar Chart for Features/Attributes')
plt.xlabel('Features/Attributes')
plt.ylabel('Total')
plt.show()
```

```python
# Q.2 Create a Scatter plot in Plotly (can be created using the scatter() method of # plotly.express.)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv')

# Create Scatter plot using Plotly
fig = px.scatter(data, x='month_number', y='total_profit', title='Scatter plot of Total Profit over Months')
fig.show()
```

```python
# Q.3 Create a Bar Chart in Plotly( can be created using the bar() method of # plotly.express class.)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv')

# Create Bar Chart using Plotly
fig = px.bar(data.drop('month_number', axis=1).sum(), title='Bar Chart of Features/Attributes')
fig.update_xaxes(title_text='Features/Attributes')
fig.update_yaxes(title_text='Total')
fig.show()
```

```python
# Q.4 Create a histograms in plotly (histograms can be created using the histogram() function of the plotly.express class.)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv')

# Create histograms using Plotly
fig = px.histogram(data, x='total_units', title='Histogram of Total Units')
fig.update_traces(marker=dict(line=dict(color='black', width=1)))
fig.show()
```
