```python
# 1. Find a dataset in the form of (x,y) and then apply K-Mean clustering.

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Importing Dataset
data = pd.read_csv('/content/sample_data/gender_submission (1).csv')
print(data)

# Selecting features and target variable
x = data[['PassengerId']]
y = data[['Survived']]

# Applying K-Means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(x)

# Plotting the clusters
plt.scatter(x.values, y.values, c=kmeans.labels_)
plt.title('K-Means Clustering')
plt.xlabel('PassengerId')
plt.ylabel('Survived')
plt.show()
```
