```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import io
import seaborn as sns

data=pd.read_csv("/content/company_sales_data (2).csv")
data

plt.bar(data["month_number"],data["facecream"],label="facecream", color ='red')
plt.bar(data["month_number"],data["facewash"],label="facewash", color='yellow')
plt.xlabel('month')
plt.ylabel('sales')
plt.title("Bar chart")
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
data['total_profit'].plot(kind='hist', bins=5, title='Total Profit Histogram', edgecolor ='black',color = 'pink')
plt.xlabel('Total Profit Range')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
