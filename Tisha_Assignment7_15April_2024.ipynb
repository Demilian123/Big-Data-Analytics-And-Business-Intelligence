```python
#Q1) Implement Simple Linear Regression (SLR) in Python using ml.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression

x=[3,5,8,9]
y=[4,6,8,6]
x=np.array([3,5,8,9]).reshape(-1,1)
y=np.array([4,6,8,6]).reshape(-1,1)

lr=LinearRegression()
lr.fit(x,y)

x_new=12
x_new = np.array(x_new).reshape(-1, 1)
y_pred=lr.predict(x_new)
print(f"Predicted value of y for x = {x_new}: {y_pred}")

import matplotlib.pyplot as plt
# Create a scatter plot of the training data
plt.scatter(x, y, color='blue', label='Training data')

# Plot the regression line
plt.plot(x, lr.predict(x), color='red', linewidth=3, label='Regression line')

# Plot the prediction for the new data point
plt.scatter(x_new, y_pred, color='green', marker='x', label='Prediction')

# Add labels and title to the plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
#Q1) Implement Simple Linear Regression (SLR) in Python using ml.

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = [2,3,6,7]
y = [3,6,4,8]
sum1 = sum(x)
sum2 = sum(y)

a = 0
for i in x:
  a += i*i

product = 0
for i in range(len(x)):
  product += x[i] * y[i]

b = len(x) * product - sum1 * sum2
b1 = len(x) * a - (sum1 * sum1)

total = b/b1
print(total)

a = sum2 * a - sum1 * product
total1 = a/b1
print(total1)
print("Linear regression equation: y =", total1, "+", total, "x")

regression_line = [total1 + total * xi for xi in x]
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, regression_line, color='red', label='Linear Regression')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()


plt.grid(True)
plt.show()
```
