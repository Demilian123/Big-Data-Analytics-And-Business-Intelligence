```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Read Total profit of all months and show it using a line plot.

# Load the dataset
df = pd.read_csv('/content/sample_data/company_sales_data (1).csv')

# Plotting data
plt.figure(figsize=(10, 6))
plt.plot(df['month_number'], df['total_profit'])
plt.title('Total Profit Over Months')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 2. Read all product sales data and show it using a multiline plot.

# Load the dataset
data = pd.read_csv('/content/sample_data/company_sales_data (1).csv')

# Plotting data
plt.figure(figsize=(10, 6))
for product in data.columns[1:7]:
          plt.plot(data['month_number'], data[product], marker='o', label=product)
plt.title('Product Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 3. Plot bar chart for all the features/attributes.

# Load the dataset
data = pd.read_csv('/content/sample_data/company_sales_data (1).csv')

# Bar chart for all attributes
plt.figure(figsize=(10, 6))
data.drop(['month_number'], axis=1).sum().plot(kind='bar')
plt.title('Bar Chart for Features/Attributes')
plt.xlabel('Features/Attributes')
plt.ylabel('Total')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  4. Plot scatter plot for 4 attributes (starting 4 attributes) with different colours.

# Load the dataset
data = pd.read_csv('/content/sample_data/company_sales_data (1).csv')

# Scatter plot for the first 4 attributes
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='month_number', y='facecream', label='Face Cream')
sns.scatterplot(data=data, x='month_number', y='facewash', label='Face Wash')
sns.scatterplot(data=data, x='month_number', y='toothpaste', label='Toothpaste')
sns.scatterplot(data=data, x='month_number', y='bathingsoap', label='Bathing Soap')
plt.title('Scatter Plot for 4 Attributes')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```
